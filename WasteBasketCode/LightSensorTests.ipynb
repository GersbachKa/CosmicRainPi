```python
#Light sensor on pin 15

import RPi.GPIO as GPIO
import time
import os
from PiPocketGeiger import RadiationWatch

pinnum = 22
```

```python
GPIO.setmode(GPIO.)
GPIO.setup(pinnum, GPIO.IN)
```

```python
total = 0
def inp(PinVal):
    global total
    x = GPIO.input(PinVal)
    if(x==0):
        total=0
        return total
    total+=1
    return total
```

```python
total=0
while(True):
    print(inp(15))
    time.sleep(.1)
    
```

```python
radWatch = RadiationWatch(18,16,numbering=GPIO.BOARD).setup()
```

```python
while True:
    y = radWatch.status()['cpm']
    print(y)
    time.sleep(.5)
```
